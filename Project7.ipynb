## 1a- Propositions

There are going to be three propositions I am going to describe below. The three propostions are going to be about choosing toppings for a pizza. Obviously pizza has the cheese and tomato sauce as the first toppings, so I am going to pick three different toppings to add on to the pizza. 

I am going to assign p, q, and r to represent these toppings. The three toppings are:

* p= pepperoni
* q= sausage
* r= pineapple 




```
AddPepperoni = True 
AddSausage = True
AddPineapple = False 
```

## 1b- The Truth of the Statements

The first statement we are going to do is $$ r \land p$$
which means "Add pineapple and add pepperoni."


```
AddPineapple and AddPepperoni
```

The next statement is going to be $$ r \lor \lnot q$$
which is "Add pineapple and do not add sausage."

```
AddPineapple or not AddSausage
```

The last truth statement I am going to describe is $$ r \Rightarrow p$$ 
This is the same as $$ \lnot r \lor p$$
This statement means "We are not going to add pineapple or we are going to add pepperoni."

```
not(AddPineapple) or AddPepperoni
```

## 1c- Truth Tables

Now I am going to develop the truth table of the statement $$ \lnot (r \oplus p)  (p \lor q)$$ 

First we need to install a truth table generator.

```
pip install truth-table-generator
```

```
import ttg
table = ttg.Truths(['p', 'q', 'r'], ['~q', 'r and p', 'r or ~q', 'r => p'], ints = False)
table.as_pandas().style.set_properties(**{'text-align': 'left'}).hide_index()
```

Now that we have the truth table generator in and we developed the truth table above, we will do the truth table of $$ \lnot (r \oplus p)  (p \lor q)$$

I am going to use XOR to put in place of $\oplus$.

```
table = ttg.Truths(['p', 'q', 'r'], ['p xor q', '~(p xor q)', 'q or r', '(~(p xor q)) => (q or r)'], ints = False)
table.as_pandas().style.set_properties(**{'text-align': 'left'}).hide_index()
```
